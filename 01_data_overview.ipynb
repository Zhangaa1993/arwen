```python
# ===== 导入库 =====
import pandas as pd
import numpy as np
import os

# 配置：显示所有列，避免截断
# show all columns
pd.set_option('display.max_columns', None)

# show up to 50 rows
pd.set_option('display.max_rows', 50)

# Set the maximum width of pandas output to 200 characters. The default is 80.
pd.set_option('display.width', 200)

# Set floating-point numbers to display with 2 decimal places.
pd.set_option('display.float_format', '{:.2f}'.format)

print('✅ 库加载完成')
```

```python
# ===== 加载所有CSV文件 =====
data_dir = 'archive/'  # 数据集所在文件夹

# 用一个字典存储所有表，方便调用
dfs = {}

file_names = {
    'orders':    'olist_orders_dataset.csv',
    'items':     'olist_order_items_dataset.csv',
    'customers': 'olist_customers_dataset.csv',
    'products':  'olist_products_dataset.csv',
    'sellers':   'olist_sellers_dataset.csv',
    'payments':  'olist_order_payments_dataset.csv',
    'reviews':   'olist_order_reviews_dataset.csv',
    'geolocation':'olist_geolocation_dataset.csv',
    'category_translation': 'product_category_name_translation.csv'
}

for key, fname in file_names.items():
    path = os.path.join(data_dir, fname)
    dfs[key] = pd.read_csv(path)
    print(f'📊 {key:20s} → {dfs[key].shape[0]:>7,} 行 × {dfs[key].shape[1]:>3} 列')

print(f'\n✅ 共加载 {len(dfs)} 张表')
```

```python
# ===== 每张表的空值比例 =====
for key, df in dfs.items():
    null_counts = df.isnull().sum()
    null_cols = null_counts[null_counts > 0]
    if len(null_cols) > 0:
        print(f'\n🔴 {key} 有空值：')
        for col, cnt in null_cols.items():
            pct = cnt / len(df) * 100
            print(f'   {col:35s} → {cnt:>8,} 个空值（{pct:.1f}%）')

print('\n✅ 空值检查完成')
```

```python
# ===== orders 表前5行 =====
orders = dfs['orders'].copy()
orders.head()
```

```python
# ===== 字段类型和含义 =====
print('📋 orders 表字段：\n')
for col in orders.columns:
    dtype = orders[col].dtype
    sample = str(orders[col].dropna().iloc[0]) if orders[col].notna().any() else 'N/A'
    print(f'  {col:35s} | {str(dtype):10s} | 样例: {sample[:50]}')
```

```python
# ===== 把时间列转成 datetime =====
time_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in time_cols:
    orders[col] = pd.to_datetime(orders[col])

print('📅 数据时间范围：')
print(f'  最早下单：{orders["order_purchase_timestamp"].min()}')
print(f'  最晚下单：{orders["order_purchase_timestamp"].max()}')
print(f'  跨度：{(orders["order_purchase_timestamp"].max() - orders["order_purchase_timestamp"].min()).days} 天')
```

```python
# ===== 订单状态分布 =====
print('📊 订单状态分布：\n')
status_counts = orders['order_status'].value_counts()

for status, cnt in status_counts.items():
    pct = cnt / len(orders) * 100
    bar = '█' * int(pct / 2)
    print(f'  {status:20s} {cnt:>7,} 单（{pct:5.1f}%） {bar}')
```

```python
# ===== 只分析已送达的订单 =====
delivered = orders[orders['order_status'] == 'delivered'].copy()
print(f'已送达订单：{len(delivered):,} 单（{len(delivered)/len(orders)*100:.1f}%）')

# 计算实际配送天数
delivered['actual_delivery_days'] = (
    delivered['order_delivered_customer_date'] - 
    delivered['order_delivered_carrier_date']
).dt.days

# 计算预计配送天数
delivered['estimated_delivery_days'] = (
    delivered['order_estimated_delivery_date'] - 
    delivered['order_delivered_carrier_date']
).dt.days

# 是否延迟
delivered['is_delayed'] = (
    delivered['order_delivered_customer_date'] > 
    delivered['order_estimated_delivery_date']
)

print(f'\n📊 配送天数统计：')
print(delivered['actual_delivery_days'].describe())
```

```python
# ===== 延迟率 =====
delay_rate = delivered['is_delayed'].mean() * 100
total = len(delivered)
delayed_count = delivered['is_delayed'].sum()

print(f'\n🚨 延迟分析：')
print(f'  总订单：{total:,} 单')
print(f'  延迟订单：{delayed_count:,} 单')
print(f'  延迟率：{delay_rate:.1f}%')
print(f'\n📊 延迟订单的配送天数分布：')
print(delivered.loc[delivered['is_delayed'], 'actual_delivery_days'].describe())
```
